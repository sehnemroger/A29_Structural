Como sa condições criticas de voo são a condição de estol e a velocidade máxima de cruzeiro, essas condições foram utilizadas no cálculo dos coeficientes aerodinâmicos. Note que, na condição de estol a aeronave está na chamada "codição de máximo arrasto", pois está com os flaps defletidos na posição máxima e está com armamento externo instalado. Já na velocide máxima de cruzeiro, chamamos de "aeonave limpa", com superficies de controle em posição neutra e sem armamento externo.


Uma informação necessária para para o cálculo dos coeficientes é o número de Reynolds na condição, dado por
$$
Re = \frac{\rho V \bar{c}}{\mu}
$$

onde $\bar{c}$ é a corda média aerodinâmica e $\mu$ é a viscosidade dinâmica do fluido ($\mu$ = 1.4207$\times$10$ ^{-5}$m$^2$/s para o ar). Para asas trapezoidais, a corda média aerodinâmica pode ser aproximada pela corda média geométrica.

Para o cálculo, foram consideradas a velocidade máxima de cruzeiro e a velocidade e condição de estol. Como o estol geralmente ocorre na altitude mínima (nível no mar), essa altitude foi adotada. Como a altitude para a velocidade máxima de cruzeiro não é conhecida, a altitude equivalente a aproximadamente 30% do teto de serviço foi adotada (3000km).

```python
from pyaat.constants import GRAVITY, RHO_SEA,DYNAMIC_VISC
from pyaat.atmosphere import atmosCOESA

atm = atmosCOESA()
atm._altitude = 3000.
rho = atm._rho

cr = 2.5 #Corda na raiz [m]
ct = 1.25 # corda na ponta [m]
l = 11.14 # envergadura [m]
V_max = 520/3.6 #Velocidade máxima de cruzeiro [m/s]
V_stall = 148/3.6 #Velocidade de estol [m/s]
cbar = (cr+ct)/2
S = cbar*l # Area em planta da asa [m^2]

cbar = (cr+ct)/2 # Corda média geometrica [m]

Re_Vmax = rho*V_max*cbar/DYNAMIC_VISC
Re_Stall = RHO_SEA*V_stall*cbar/DYNAMIC_VISC

print("Numero de Reynolds na velocidade máxima de cruzeiro:")
print(Re_Vmax)
print("--------------------------------------")
print("Numero de Reynolds na velocidade máxima de estoll:")
print(Re_Stall)
```

Em posse dessas Infromações, o cálculo dos coeficientes de sustentação dos prefis pode ser realizado. Para tál, utilizou-se a ferramenta XFloil, que utilza metodos potênciais para clálculo da sustentação. A ferramenta também possui correções para incluir arrasto aerodinâmico.

Realizou-se análises no range de -25$^o$ 1a +35$^o$, considerando-se o número de Reynolds de 15M. Note que os maiores ângulos de atanque não convergiram.

Os resultados para ambos os perfis são apresentados abaixo.

Na condição de estol, esses aerofólios e geometria não são representativos da real configuração da aeronave. Com os flaps defletidos tando a área em planta, quando a curvatura do perfil aumentam. Por consequência, os coefientes de sustentação e arrasto aumentam considerávelmente (uma vez que continuam sendo normalizados pelos parâmtros originais).

Uma vez não possuimos informações da nova geometria, precisamos partir de parâmetros de desempenho para estima-los. Como a condição de estol é totalemte conhecida, temos:

$$
C_{Lstoll} = \frac{2 M_{TOW} g}{S V_{stall}^2 \rho}
$$

```python
MTOW = 5400 #[kg]
CL_stall =(2*MTOW*GRAVITY/(S*V_stall**2*RHO_SEA))

print("CL na condição de estol")
print(CL_stall)
```

Como uma aproximação, consideramos que a aeronave aumente sua forma em planta em 30% quando os flaps são acionados. Dessa forma, o aerofólio equivalente na condição de estol deve fornecer uma sustentação de aproximadamente
$$
C_{Leq} = \frac{CL_Stall}{1.3}
$$

```python
C_Leq = CL_stall/1.3
print("CL equivalente")
print(C_Leq)
```

Como o aerofólio orginal possui 15% de espessura máxima em relação a corda, o novo aerofólio deve possuir um valor aproximadamente 30% menor (11.5%), dado que a maxima espessura não é alterada, mas a corda sim.

# 4 Diagrama V-n

Na análise dos limites de cargas foi adotado o regulamento MIL-A-8861B [2] e MIL-A-8860B [5]. Esse regulamento especifica requisitos estruturais (rigidez e resistência) para aeronaves militares sob diferentes condições de carregamento, incluindo condições limite de operação em atmosfera estacionária e sob condições de vento.

De acordo com a norma reguladora, a aeronave se encaixa na cadegoria "aeronave de ataque", onde os fatores de carga para manobras no plano de simetria são:

|__Maximo fator de carga (_voo básico_)__| __Mínimo fator de carga (_voo básico_)__ |__Maximo fator de carga (_MTOW_)__| __Mínimo fator de carga (_MTOW_)__ | __Fator de carga em $V_L$__|
|:----------:|:-------------:|:----------:|:-------------:|:-------------:|
|<img width=150/>|<img width=150/>|<img width=150/>|<img width=150/>|<img width=150/>|
|7.5|-3.5|5.5|-2|-1

O termo _voo basico_ não é definido pela norma, por esta razão será definico como o peso da aeronave com 60% da carga paga, que é o comumente empregado na literatura para missões de longa distância. Nesse caso, o peso da aeronave em um voo básico é de 4625kg.

Usualmente o diagrama é construido em termos da velocidade equivalente ao nível do mar e expresso em nós por segundo, porém, nesse trabalho será utilizado o sistema internacional (metros por segundo).

Do ponto de vista estrutural, apenas o caso onde ocorrem os maiores fatores de carga são de interesse, por essa razão, apenas a condição de voo básico será considerada.

## 4.1 Definições

Antes da construção do diagrama é conveniente a apresentação de algumas definições e equações.
* $V_H$ é a velocidade máxima de voo nivelado utilizando-se a pontência máxima do motor (incluindo afterburner, se disponível). Para a aeronave EMB314 esse valor é de 590km/h (163.88 m/s);

* $V_L$ é a valocidade máxima de mergulho. Como esse valor não é conhecido, seguimos a recomendação de [6] e adotamos $V_L = 1.25V_H$.

* $V_{EAS}$ é a velocidade equivalente à nível do mar. Obtida multiplicando-se a velocidade aerodinâmica $V_{TAS}$ por um fator de correção. $V_{EAS} = V_{TAS} \sqrt{\frac{\rho}{\rho_0}}$.

* $n_{z+}$ é o fator de carga normal positivo, dado pela expressão

$$
n_{z+} = K\frac{\rho V_e^2S}{2 (M_{TO})g}C_{nMax}
$$

em que $K = 1.25$ para aeronaves em velocidades $M \leq 0.6$ e $C_{nMax}$ é o coeficiente de força normal, dado por:

$$
C_{nMax} = \sqrt{C_{Lmax}^2 + C_{DaMax}^2}
$$

onde $C_{DaMax}$ é o coeficiente de arrasto no ângulo de $C_{Lmax}$.

$V_e$, por sua vez, é a menor velocidade onde a aeronave gera o fator de carga de design quando aplicadas as deflexões máximas nas superfícies de controle em uma manobra de pull-out.

* $n_{z-}$ é o fator de carga normal negativo. Equivalente ao $n_{z+}$ porém para manobras negativas.

Ambos $n_{z-}$ e $n_{z+}$ são calculados na codndição de máximo arrasto.

```python
from numpy import sqrt, array, arange, ones

atm._altitude = 3000.
rho3000 = atm._rho

# Fatores de carga
np_bf = 7.5
nn_bf = 3

# Cálculo dos valores constantes
K = 1.25 
MBF = 4625 #[kg]
VH = 163.88 #[m/s]
VHe = VH*sqrt(rho3000/RHO_SEA)
VL = K*VH #[m/s]
VLe = VL*sqrt(rho3000/RHO_SEA)

CDaMax_drag = results_polar_s[list(results_polar_s[:,1]).index(max(results_polar_s[:,1])),2] # Valor de CD no ponto CLmax
CDaMin_drag = results_polar_s[list(results_polar_s[:,1]).index(min(results_polar_s[:,1])),2] # Valor de CD no ponto CLmin
CnMax = sqrt(CLmax_drag**2+CDaMax_drag**2) #Coeficiente de força vertical positivo
CnMin = sqrt(CLmin_drag**2+CDaMin_drag**2) #Coeficiente de força vertical negativo

# Cálculo da velocidade minima de manobra no fator de carga máximo
Ve_pos = sqrt(2*np_bf*MBF*GRAVITY/(RHO_SEA*K*CnMax*S))
Ve_pos_min = sqrt(2*1*MBF*GRAVITY/(RHO_SEA*K*CnMax*S))
Ve_neg = sqrt(2*nn_bf*MBF*GRAVITY/(RHO_SEA*K*CnMin*S))
Ve_neg_min = sqrt(2*1*MBF*GRAVITY/(RHO_SEA*K*CnMin*S))

print("Velocidade onde nz = 7.5:", Ve_pos)
print("Velocidade onde nz = -3.5:",Ve_neg)

VlineY = arange(0, 1, 0.01)
onesV = ones(len(VlineY))

zeroline_x = arange(0, VLe, 0.1)
zeroline_y = zeroline_x[:]*0.0

Ve_pos_array = arange(Ve_pos_min, Ve_pos, 0.5)
Ve_neg_array = arange(Ve_neg_min, Ve_neg, 0.5)

Ve_pos_array_stall = arange(0, Ve_pos_min, 0.5)
Ve_neg_array_stall = arange(0, Ve_neg_min, 0.5)

OA = K*RHO_SEA*S*CnMax/(2*MBF*GRAVITY)*Ve_pos_array**2
OE = -K*RHO_SEA*S*CnMin/(2*MBF*GRAVITY)*Ve_neg_array**2

OA_min = K*RHO_SEA*S*CnMax/(2*MBF*GRAVITY)*Ve_pos_array_stall**2
OE_min = -K*RHO_SEA*S*CnMin/(2*MBF*GRAVITY)*Ve_neg_array_stall**2

AB_x = arange(Ve_pos, VLe, 0.5)
AB_y = np_bf*ones(len(AB_x))

ED_x = arange(Ve_neg, VHe, 0.5)
ED_y = -nn_bf*ones(len(ED_x))

BC_y = arange(-1, 7.5, 0.01)
BC_x = VLe*ones(len(BC_y))

DC_x = arange(VHe, VLe, 0.5)
DC_y = -nn_bf + (DC_x - VHe)*((+nn_bf - 1)/(VLe - VHe))

print(VLe)
```

### Diagrama de rajadas

A norma preve que a aeronave suporte cargas de vento causadas por ventos verticais e laterais. Essas cargas são especificadas em termos de rajadas discretas e regimes turbulêntos contínuos. Nesse trabalho serão analizadas apenas as condições de rajadas discretas.

A normas especifica 4 condições distintas, sendo elas:
* A) 20.12 m/s e $V_G$

* B) 15.24 m/s em $V_H$

* C) 7.62 m/s em $V_L$

* D) 15.24 m/s em $V_{LH}$ (Condição de decolagem e pouso)

A condição D não será analizada pois requer uma análise aerodinâmica mais detalhada da aeronave completa.

Para a condição de rajadas considera-se apenas a sitação onde a aeronave está com os flaps recolhidos.

O fator de carga da aeronave quando sujeita a rajadas varia linearmente de 1 até o valor máximo, desctito pela expressão

$$
n = 1 + \frac{K_W V_e U_{de} C_{n \alpha} S}{498 M_{TO}}
$$

onde $K_W$ é um fator de rajada, $U_{de}$ é a velocidade máxima equivalente do vento, e $C_{n \alpha}$ é a derivada do fator de carga $C_n$ em função do ângulo de ataque $\alpha$ (em radianos).

Para o caso subsônico, $K_W$ é dado pela expressão
$$
K_W = \frac{0.88 \mu}{5.3 + \mu}
$$
onde $\mu$ é uma razão de massa, dada por
$$
\mu = \frac{2 M_{TO}}{\rho S \bar{c} C_{L \alpha}}
$$
onde $C_{L \alpha}$ é a inclinação da curva $C_L$ em função de $\alpha$.

Note que todas as unidades de massa, área e velocidade devem ser convertidas para o sistema britânico para uso das expressões.

A curva de $C_{n}$ em função de $\alpha$ pode ser construida a partir das curvas de $C_L$ e $C_D$. Como segue.

```python
cn = []
for i in range(0, len(results_polar[:,1])):
    if results_polar[i,1]<=0:
        cn.append(-sqrt(results_polar[i,1]**2+ results_polar[i,2]**2))
    else:
        cn.append(sqrt(results_polar[i,1]**2+ results_polar[i,2]**2))
        
plt.figure(figsize=(14, 3), dpi=300)
plt.plot(results_polar[:,0],cn,color = 'red')
plt.title(r"Curva $C_n- \alpha$ da asa da aeronave A29")
plt.xlabel(r"Ângulo de ataque [$^o$]")
plt.ylabel("Coeficiente de fator de carga [adm]")
plt.grid()
```

Note que uma adaptação dever er feia na equação para extrair o sinal do fator de carga, pois o valor gerado orginalmente é exclusivamente positivo.

A derivada é computada entre os ângulos -10$^o$ e +10$^o$, como segue

```python
from numpy import radians
Cna = (cn[303]- cn[103])/(radians(results_polar[303,0] - results_polar[103,0]))
print("Cn-alpha:", Cna)
```

O mesmo procedimento é adotado para o cálculo de $C_{L \alpha}$, como segue.

```python
CLa = (results_polar[303,1]- results_polar[103,1])/(radians(results_polar[303,0] - results_polar[103,0]))
print("CL-alpha:", CLa)
```

Como pode ser verificado $C_{L \alpha}$ e $C_{n \alpha}$ possuem valores muito próximos. Isso é esperado dado que o coeficiente de fator de carga depende majoritariamente do coeficiente de sustentação.

A partir disso podemos computar o coeficiente $\mu$

```python
mu = (2*MBF)/(RHO_SEA*S*cbar*CLa)
print(r"$\mu$:", mu)
```

Que nos fornece o fator de rajadas de

```python
KW = (0.88*mu)/(5.3+mu)
print("Kw:", KW)
```

```python
from numpy import linspace
from scipy.optimize import root
kg2pd = 2.2
mps2kts = 1.94384
mps2fps = 3.28084
msq2fsq = 10.7639
Ude_VH = 15.24*sqrt(rho3000/RHO_SEA)
Ude_VG = 20.12*sqrt(rho3000/RHO_SEA)
Ude_VL = 7.62*sqrt(rho3000/RHO_SEA)
VHe_x = linspace(0, VHe, 50)
VLe_x = linspace(0, VLe, 50)


VHe_y = 1 + KW*Ude_VH*mps2fps*Cna*VHe_x*mps2fps*S*msq2fsq/(498*MBF*kg2pd)
VHe_yn = 1 - KW*Ude_VH*mps2fps*Cna*VHe_x*mps2fps*S*msq2fsq/(498*MBF*kg2pd)
VLe_y = 1 + KW*Ude_VL*mps2fps*Cna*VLe_x*mps2fps*S*msq2fsq/(498*MBF*kg2pd)
VLe_yn = 1 - KW*Ude_VL*mps2fps*Cna*VLe_x*mps2fps*S*msq2fsq/(498*MBF*kg2pd)

def findVG(Ve):
    return 1 + KW*Ude_VH*mps2fps*Cna*VHe_x*mps2fps*S*msq2fsq/(498*MBF*kg2pd)-K*RHO_SEA*S*CnMax/(2*MBF*GRAVITY)*Ve_pos_array_stall**2

VG = scipy.optimize.root(findVG, VHe)
```
